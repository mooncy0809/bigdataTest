```python
# select = [1, 1, 1] # 닭가슴살 추천 필요
# select = [1, 2, 1] # 닭가슴살 추천 필요
# select = [1, 1, 2] # 닭가슴살 추천 필요
# select = [2, 1, 1] # 닭가슴살 추천 필요
# select = [3, 1, 1] # 닭가슴살 추천 필요
# select = [2, 2, 2] # 간편요리 추천 필요
# select = [1, 2, 2] # 간편요리 추천 필요
# select = [2, 1, 2] # 간편요리 추천 필요
# select = [2, 2, 1] # 간편요리 추천 필요
# select = [3, 3, 3] # 샐러드 추천 필요
# select = [1, 3, 3]   # 샐러드 추천 필요
sample = [[1, 1, 1, 1, 1],[1, 2, 1, 4, 2],[1, 1, 2, 3, 4],[2, 1, 1, 4, 4],[3, 1, 1, 4, 5], [1, 3, 1,5 ,5]]
sample = sample + [[2, 2, 2, 5, 5],[1, 2, 2, 3, 4],[2, 1, 2, 4, 4],[2, 2, 1, 5, 5]]
sample = sample + [[3, 3, 3, 5, 4], [1, 3, 3, 2, 1], [3, 1, 3, 2, 3], [3, 3, 1, 4, 4], [2, 3, 3, 4, 4], [3, 2, 3, 4, 4], [3, 3, 2,  4, 4]]
sample = sample + [[1, 2, 3,2 ,2], [3, 2, 1, 4, 4]]
for select in sample: # 2차원 list
    print(select) # 1차원 list [1, 1, 1]

    X = []
    for i in range(0, 5): # 평가 화면(횟수) 3번, 0 ~ 2
        for j in range(1, 6): # 평가 항목수 3개
            # print(i, j)
            # 선택한 번호화 같은 경우만 1을 할당
            # 사용자가 입력한 1은 배열 index 0에 해당함.
            if (select[i] == j): # 1 == 1
                X.append(0.96)   
            else:
                X.append(0.01)
            print(X, end='')    
        print()    

    data = np.array([np.array(X)]) # 입력 데이터는 ndarray 타입의 2차원이어야함, 행 방향
    p = model.predict(data)  # 모델 사용, 2차원 데이터 전달
    print(p)
    index = np.argmax(p[0])  # 가장 높은 값의 index 산출, 가장 확률이 높은 값의 index
    if index == 0:
        label = '닭가슴살 카테고리'
    elif index == 1:
        label = '간편요리 카테고리'
    elif index == 2:
        label = '샐러드 카테고리'
    elif index == 3:
        label = '건강미용 카테고리'
    elif index == 4:
        label = '간식 카테고리'
    print(label + ' 추천' +'\n') 
```
