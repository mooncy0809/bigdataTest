```python
# 5. 다음은 California Housing 데이터 세트이다. 
# 주어진 데이터의 첫 번째 행부터 순서대로 80%까지의 데이터를 훈련 데이터로 추출 후, 
# 전체 방 개수 컬럼을 의미하는 'total_bedrooms' 변수의 결측치를
# 'total_bedrooms' 변수의 중앙값으로 대체한 데이터 세트를 구성한다. 
# 결측치 대체 전의 'total_bedrooms' 변수 표준편차 값과
# 결측치 대체 후의 total_bedrooms' 변수 표준편차 값의 차이에 대한 절댓값을 계산하시오.
# answer: 1.975147291645726
df = pd.read_csv('./data/202_housing.csv')
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 주어진 데이터의 첫 번째 행부터 순서대로 80%까지의 데이터를 훈련 데이터로 추출 후, 
train,test=train_test_split(df,test_size=0.2,shuffle=False)
```

```python
train.describe()
```

```python
# 전체 방 개수 컬럼을 의미하는 'total_bedrooms' 변수의 결측치를
# 'total_bedrooms' 변수의 중앙값으로 대체한 데이터 세트를 구성한다. 
```

```python
med=train['total_bedrooms'].median()
med
```

```python
train.isna().sum()
```

```python
org_train=train.copy()
```

```python
train['total_bedrooms'].fillna(med,inplace=True)
```

```python
train.isna().sum()
```

```python
org_train.isna().sum()
```

```python
# 결측치 대체 전의 'total_bedrooms' 변수 표준편차 값과
# 결측치 대체 후의 total_bedrooms' 변수 표준편차 값의 차이에 대한 절댓값을 계산하시오.
print(abs(org_train['total_bedrooms'].std()-train['total_bedrooms'].std()))
```

```python
# 6. 2번 문항에서 활용한 California Housing 데이터 세트를 그대로 활용한다. 
# 인구 컬럼인 population 항목의 이상값의 합계를 계산하시오. 
# (※ 이상값은 평균에서 1.5 × 표준편차를 초과하거나 미만인 값의 범위로 정한다.)
# answer: 5607295.0
df = pd.read_csv('./data/203_housing.csv')
```

```python
# 인구 컬럼인 population 항목의 이상값의 합계를 계산하시오. 
```

```python
a,b=df['population'].mean()-1.5*df['population'].std(),df['population'].mean()+1.5*df['population'].std()
print(a)
print(b)
```

```python
temp=(df['population']<a)|(df['population']>b)
```
